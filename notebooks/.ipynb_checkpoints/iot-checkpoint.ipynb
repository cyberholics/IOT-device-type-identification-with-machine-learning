## Remove features with low variance(Close to 0), because they are are constant and do not add any value to the model

```python
var_thresh = VarianceThreshold(threshold=0.1)
X = df.drop(["device_category"],axis=1)
y = df["device_category"]
var_thresh.fit(X)
var_thresh.get_support()
```

```python
concol = [column for column in X.columns 
          if column not in X.columns[var_thresh.get_support()]]

for features in concol:
    print(features)
```

```python
X = X.drop(concol,axis=1)
X
```

### Perform standardization on the data


```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

### Spliting the data

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)
```

### Building model (Logistic regression)

```python
lr = LogisticRegression(max_iter=10000)
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

### Building model (Random Forest)

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

### Building model (Gradient boosting)

```python
GBC = GradientBoostingClassifier()
GBC.fit(X_train, y_train)
GBC.score(X_test, y_test)
```

### Observation: The best model for this task is the Random forest classfier

### Using grid search cv to find the best hyperparameters 

```python
# Define the hyperparameter grid to search over
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],  # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],  # Minimum number of samples required to be at a leaf node
    'bootstrap': [True, False],  # Whether to use bootstrap samples when building trees
}

rf = RandomForestClassifier(random_state=0)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best_rf_model = grid_search.best_estimator_
accuracy = best_rf_model.score(X_test, y_test)
print("Best Hyperparameters:", best_params)
print("Accuracy on Test Data:", accuracy)
```

### Training final model (Random forest model)

```python
hyperparameters = {
    'bootstrap': False,
    'max_depth': None,
    'min_samples_leaf': 1,
    'min_samples_split': 5,
    'n_estimators': 200
}

final_model = RandomForestClassifier(**hyperparameters)
final_model.fit(X_train, y_train)
predictions = final_model.predict(X_test)
predictions
```

### Evaluating final model 

#### Accuracy score

```python
accuracy_score(y_test,predictions)
```

#### Cross Validation (To check for overfitting)

```python
model = RandomForestClassifier(**hyperparameters)
k = 10
scores = cross_val_score(model, X_train, y_train, cv=k, scoring='accuracy')
mean_accuracy = scores.mean()
std_accuracy = scores.std()
print(scores)
print(f"Mean Accuracy: {mean_accuracy:.2f}")
print(f"Standard Deviation of Accuracy: {std_accuracy:.2f}")
```

#### Classification report 

```python
print(classification_report(predictions,y_test))
```
